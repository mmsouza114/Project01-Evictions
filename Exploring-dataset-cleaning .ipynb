# Exploring the dataset and cleaning it up

### First let's import pandas API

```python
import pandas as pd
```

```python
evictions_df = pd.read_csv("Evictions.csv")
#evictions_df = pd.read_csv('Evictions.csv', index_col=0, parse_dates=[4])

#this will show me the first 10 rows in the data sheet
evictions_df.head(10)
```

```python
#Show a list of columns for easier refernce to check which ones I am interested in
evictions_df.info()
```

```python
#Now to delete the ones I am not interested in 
#First lets use the columns function to make it easier to delete the ones I dont need
evictions_df.columns
```

```python
evictions_df_subset= evictions_df[['Court Index Number', 'Docket Number ', 'Eviction Address',
       'Eviction Apartment Number', 'Executed Date', 'Marshal First Name',
       'Marshal Last Name', 'Residential/Commercial', 'BOROUGH',
       'Eviction Postcode', 'Ejectment', 'Eviction/Legal Possession',
       'Latitude', 'Longitude', 'Community Board', 'Council District',
       'Census Tract', 'BIN', 'BBL', 'NTA']]
evictions_df_subset
```

```python
#Now lets replace the spaces between the column names with "_" so it's easier to work with
evictions_df_subset.columns = evictions_df_subset.columns.str.replace(' ','_')


evictions_df_subset.info()
```

```python
#convert date executed type object into datetime
evictions_df_subset['Executed_Date']= pd.to_datetime(evictions_df_subset['Executed_Date'])

#call info function again to double check it changed type
evictions_df_subset.info()
```

```python
#Now let us see if there are any null values.
evictions_df_subset.isna().sum()
```

```python
evictions_df_subset[evictions_df_subset['Latitude'].isna()]
```

## My review of the nulls led to bringing back the deleted rows

Intially I removed some rows (Community_Board, Council_District, Census_Tract, BIN, BBL, NTA) because I didnt think that I would need them in analysising this data set, but when I looked at the nulls in the Latitude and 	Longitude, it didn't make sense why those are null only with the rows I had left, so I add the rest of the deleted rows back and I noticed that those rows were null too.

```python
#lets find what BIN and BBL are null but the other rows aren't 
pd.set_option("display.max_rows", None)
evictions_df_subset.loc[(evictions_df_subset['BIN'].isna())
                        & (evictions_df_subset['BBL'].isna())
                        & (evictions_df_subset['Latitude'].notna())]
```
